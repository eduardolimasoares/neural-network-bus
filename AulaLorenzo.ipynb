```python
numero_um = 12
numero_dois = 32
```

```python
# operadores aritiméticos
print(numero_um + numero_dois)
print(numero_um * numero_dois)
print(numero_um - numero_dois)
print(numero_um / numero_dois)
print(2**2)
print(2//2)
```

```python
# Operadores Relacionais
# O resultado de uma operaç~~ao relacional sempre é um, um, um BOOOLEEEAAAANNOOOOOOO
print(10 < 5)
print(10 > 5)
print(10 == 5)
print(10 != 5)
print((10 > 5) and (10 < 5))
print((10 == 5) or (10 < 5))
```

```python
import http.server
import socketserver
import requests
from dataclasses import dataclass
from typing import List
import json

@dataclass
class Post:
    UserId: int
    Id: int
    Title: str
    Body: str

@dataclass
class Posts:
    Posts: List[Post]



class MyRequestHandler(http.server.SimpleHTTPRequestHandler):
    r = requests.get('https://jsonplaceholder.typicode.com/posts')
    posts = Posts(r.json())

    def do_GET(self):
        if self.path == '/':
            self.path = '/index.html'
        if self.path == '/otario':
            self.path = '/otario.html'
        return http.server.SimpleHTTPRequestHandler.do_GET(self)


Handler = MyRequestHandler
server = socketserver.TCPServer(('localhost', 8080), Handler)

server.serve_forever()
```
